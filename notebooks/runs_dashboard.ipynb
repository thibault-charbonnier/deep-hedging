```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

# Trouve la racine projet même si le notebook est lancé depuis notebooks/
cwd = Path.cwd()
if (cwd / "outputs").exists():
    project_root = cwd
elif (cwd.parent / "outputs").exists():
    project_root = cwd.parent
else:
    raise FileNotFoundError(f"Impossible de trouver le dossier outputs depuis {cwd}")

outputs = project_root / "outputs"
print("Project root:", project_root)
print("Outputs dir :", outputs)

candidates = sorted(
    outputs.glob("*/data/train_steps.csv"),
    key=lambda p: p.stat().st_mtime,
    reverse=True,
)

if not candidates:
    raise FileNotFoundError(f"Aucun train_steps.csv trouvé dans {outputs}/*/data/")

train_csv = candidates[0]
print("Using:", train_csv)

df = pd.read_csv(train_csv)
loss = df["loss"].dropna().reset_index(drop=True)

plt.figure(figsize=(8, 4))
plt.plot(loss, alpha=0.3, label="raw")
plt.plot(loss.rolling(200, min_periods=1).mean(), linewidth=2, label="rolling mean (200)")
plt.title(f"Training loss - {train_csv.parent.parent.name}")
plt.xlabel("Update index")
plt.ylabel("Loss")
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```
